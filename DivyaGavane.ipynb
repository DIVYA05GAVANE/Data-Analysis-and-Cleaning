```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('googleplaystore.csv')
df
```

```python
df['Installs']=df['Installs'].replace('Free',np.nan)
df.dropna(inplace=True)
df
```

```python
#"Category having Maximum number of apps"
```

```python
category = df['Category']
category
category = category.unique()
category
```

```python
dup_category = df.pivot_table(index=['Category'],aggfunc='size')
print(dup_category)
maximum=dup_category.idxmax()
```

```python
#"ANSWER"
```

```python
print(maximum)
```

```python
data_top=dup_category.head(33)
list(data_top.index.values.tolist()) 
```

```python
plt.bar(category.transpose(),dup_category,color='green')
plt.title("App categories")
plt.xlabel("Category")
plt.ylabel("No. of Apps")
plt.xticks(category,list(data_top.index.values.tolist()) ,rotation=90)
plt.show()
```

```python
#"Top 5 largest apps in given dataset having installs greater than 1 million"
```

```python
df1=df
size=df1['Size']
size=size.replace('k',np.nan,regex=True)
size.dropna(inplace=True)
size=size.replace('M','',regex=True)
df1['Size']=size
df1['Size']=df1['Size'].replace('Varies with device',np.nan)
df1.dropna(inplace=True)
df1
```

```python
df1=df1[['App','Size','Installs']]
df1=df1[df1['Installs']>1000000]
df1['Size']=df1['Size'].astype(float)
df1=df1.sort_values(by='Size',ascending=False)
#df1=df1.sort_values(by='Installs',ascending=False)
df1
```

```python
#"ANSWER"
```

```python
df1.nlargest(5,['Size'])
```

```python
#"How App ratings differ between paid and free apps in general?"
```

```python
df3=df[['Rating','Price']]
df3
```

```python
df4=df3[df3['Price']=='0']
df4
```

```python
ndf4 = df4.pivot_table(index=['Rating'],aggfunc='size')
ndf4=ndf4.sort_index(ascending=False)
ndf4
```

```python
df5=df3[df3['Price']!='0']
df5
```

```python
ndf5 = df5.pivot_table(index=['Rating'],aggfunc='size')
ndf5=ndf5.sort_index(ascending=False)
ndf5
```

```python
#"There are 236 free apps with rating 5.0, while there is only 25 paid apps with the same rating."
#"The pattern seen from the two datasets namely 'ndf4' and 'ndf5' is that there are more number of free apps for a particular rating as compared to that of paid apps." 
```
